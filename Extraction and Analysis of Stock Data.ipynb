```python
from datetime import date
```

```python
# create a list of all stock you want to extract their data
symbol1 = ["NIFTY SMLCAP 100", "NIFTY SMLCAP 250", "NIFTY SMLCAP 50",
          "NIFTY SMLCAP 100", "NIFTY MIDCAP 50", "NIFTY 50", "NIFTY NEXT 50", "NIFTY FMCG"]

## create a dummy box
data1= []
data1 = pd.DataFrame(data1)
```

```python
for x in symbol1:
    data = get_history(symbol=x, start=date(2019, 1, 1), end=date(2020, 12, 31), index = True)
    data= pd.DataFrame(data)
    data["indexname"] = x
    data1 = pd.concat([data1,data])
    print(x)
```
